#1. CREATE A DATAFRAME


```python
import pandas as pd
import numpy as np
from functools import reduce



try:
    # Creating a DataFrame with Name, Age, and City
    data = {
        "Name": ["Alice", "Bob", "Charlie"],
        "Age": [25, 30, 35],
        "City": ["Toronto", "Ottawa", "Montreal"]
    }

    df = pd.DataFrame(data)

    print("Original DataFrame:")
    print(df)
    print("\nData types:")
    print(df.dtypes)

except Exception as e:
    print("Error when creating DataFrame:", e)
```

#4. GROUPBY AND DESCRIBE

```python
try:
    df_group = pd.DataFrame({
        "Category": ["A", "A", "B", "B", "B"],
        "Value": [10, 20, 30, 40, 50]
    })

    grouped_stats = df_group.groupby("Category")["Value"].describe()

    print("Grouped statistics:")
    print(grouped_stats)
    print()

except Exception as e:
    print("Error in GroupBy or describe:", e)
```

#5. CONCATENATION AND MERGING

```python
try:
    df1 = pd.DataFrame({
        "ID": [1, 2],
        "Name": ["Alice", "Bob"]
    })

    df2 = pd.DataFrame({
        "ID": [3, 4],
        "Name": ["Charlie", "David"]
    })

    df3 = pd.DataFrame({
        "Age": [25, 30, 35, 40],
        "City": ["Toronto", "Ottawa", "Montreal", "Calgary"]
    })

    # Vertical concatenation
    concat_df = pd.concat([df1, df2], axis=0, ignore_index=True)

    # Horizontal merge by index using concat
    merged_df = pd.concat([concat_df, df3], axis=1)

    print("Concatenated DataFrame (vertical):")
    print(concat_df)
    print()

    print("Merged DataFrame (horizontal):")
    print(merged_df)
    print()

    print("\nFinal shape:")
    print(merged_df.shape)
    print()

except Exception as e:
    print("Error in concatenation or merging:", e)
```

#9. ITERATORS AND GENERATORS

```python
try:
    class EvenIterator:
        def __init__(self):
            self.current = 0
            self.count = 0

        def __iter__(self):
            return self

        def __next__(self):
            if self.count < 5:
                result = self.current
                self.current += 2
                self.count += 1
                return result
            else:
                raise StopIteration

    def even_generator():
        for i in range(0, 10, 2):
            yield i

    print("Output with Iterator:")
    even_iter = EvenIterator()
    for num in even_iter:
        print(num, end=" ")
    print()

    print("Output with Generator:")
    for num in even_generator():
        print(num, end=" ")
    print()

except Exception as e:
    print("Error with iterator or generator:", e)
```

#10. MAP, REDUCE, AND FILTER

```python
try:
    nums = [1, 2, 3, 4, 5]

    squared = list(map(lambda x: x ** 2, nums))
    product = reduce(lambda x, y: x * y, nums)
    evens = list(filter(lambda x: x % 2 == 0, nums))

    print("Original list:", nums)
    print("Squared with map:", squared)
    print("Product with reduce:", product)
    print("Even numbers with filter:", evens)

except Exception as e:
    print("Error with map, reduce or filter:", e)
```

#11. OBJECT-ORIENTED PROGRAMMING

```python
try:
    class Rectangle:
        def __init__(self, length, width):
            if length <= 0 or width <= 0:
                raise ValueError("Length and width must be positive numbers.")
            self.length = length
            self.width = width

        def area(self):
            return self.length * self.width

        def perimeter(self):
            return 2 * (self.length + self.width)

    rect1 = Rectangle(5, 3)
    rect2 = Rectangle(10, 4)

    print("Rectangle 1:")
    print("Length =", rect1.length, "Width =", rect1.width)
    print("Area =", rect1.area())
    print("Perimeter =", rect1.perimeter())

    print("\nRectangle 2:")
    print("Length =", rect2.length, "Width =", rect2.width)
    print("Area =", rect2.area())
    print("Perimeter =", rect2.perimeter())

except ValueError as ve:
    print("Value error in Rectangle class:", ve)
except Exception as e:
    print("Error with class:", e)
```

#12. PANDAS DATA ANALYSIS

```python
try:
    employees = pd.DataFrame({
        "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
        "Department": ["HR", "IT", "IT", "Finance", "HR"],
        "Salary": [55000, 70000, 65000, 80000, 60000]
    })

    print("Employees DataFrame:")
    print(employees)
    print()

    # Average salary by department
    avg_salary = employees.groupby("Department")["Salary"].mean()

    print("Average salary by department:")
    print(avg_salary)
    print()

    # Employees with salary > 60000
    high_salary_names = employees.loc[employees["Salary"] > 60000, "Name"]

    print("Employees with salary > 60000:")
    print(list(high_salary_names))
    print()


    # Add Bonus column
    employees["Salary Bonus"] = employees["Salary"] * 0.10

    print("\nEmployees DataFrame after adding Bonus:")
    print(employees)
    print()

except Exception as e:
    print("Error in pandas data analysis:", e)
```
