# GENERATIVE AI

# Install libraries

```python
# 1. Install Libraries
!pip install transformers torch
# 2. Import Libraries
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

# Load Model and Tokenizer

```python

model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()
```

#Function to generate Text

```python
def generate_text(prompt, max_length=100, temperature=0.7, top_k=50, top_p=0.95):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(
        inputs,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True
    )
    text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return text
```

# Basic Test Generation

```python

prompt1 = "Once upon a time"
print(generate_text(prompt1))
```

# Output Length Control

```python

prompt2 = "The future of artificial intelligence is"
print("\nShort Output:\n", generate_text(prompt2, max_length=50))
print("\nLong Output:\n", generate_text(prompt2, max_length=200))
```

# Temperature Control

```python

prompt3 = "The benefits of renewable energy include"
print("\nLow Temperature (less creative):\n", generate_text(prompt3, temperature=0.2))
print("\nHigh Temperature (more creative):\n", generate_text(prompt3, temperature=1.5))
```

# Top_k and Top_p

```python
prompt4 = "In the year 2050, transportation will"
print("\nDefault:\n", generate_text(prompt4))
print("\nRestricted Output:\n", generate_text(prompt4, top_k=10, top_p=0.5))
```

# Prompt Engineering

```python
print("\n===== Exercise 5: Prompt Engineering =====")
base_prompt = "Write a short story about"
topics = ["a robot teacher", "a magical village", "an AI-powered world"]
for topic in topics:
    full_prompt = f"{base_prompt} {topic}"
    print("\nPrompt:", full_prompt)
    print(generate_text(full_prompt, max_length=150))
```

# Interactive Prompt

```python
user_input = input("Enter your prompt: ")
print(generate_text(user_input))
```

# Chat-like Interaction

```python
def ask_gpt2(prompt):
    return generate_text(prompt, max_length=100)
system_prompt = "You are a helpful assistant."
user_prompt = "Explain machine learning in simple terms."
full_prompt = system_prompt + "\nUser: " + user_prompt + "\nAI:"
print(ask_gpt2(full_prompt))
```

# Multi-turn Chat

```python
def chat_with_gpt():
    print("Type 'exit' to stop")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            break
        response = generate_text(user_input)
        print("AI:", response)
chat_with_gpt()
```
